```python

def to_camel_case(text):
    text = text.replace("-", " ").replace("_", " ")
    words = text.split()
    return "".join([w.capitalize() if w != words[0] else w for w in words])

C1=("the-stealth-warrior")
C2=("The-Stealth-Warrior")
C3=("The_Stealth-Warrior")
print(to_camel_case(C1))
print(to_camel_case(C2))
print(to_camel_case(C3))
```
